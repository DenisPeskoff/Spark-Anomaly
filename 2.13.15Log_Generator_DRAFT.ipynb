```python
#this program generates log files with a 1% anomaly rate

import random
import socket
import struct
import time
import datetime

#Global Variable time sets a starting date
#time wll be split into date and time
Time = datetime.datetime(2000, 2, 3, 4, 5, 6) 


#global locks that will be used for anomaly scenarios
lockTime = 0
lockIP = 0
lockAction = 0
lockActionStatus = 0
lockPort = 0
lockUser = 0

#stores last port and IP used for anomaly scenarios
lastPort = 0
lastIP = 0
```

```python
#this function decides if a normal line or an Anomaly will be generated

def createLogFile():
    
    file = open("logfile.log", "a") #the file that will be written
    
    for i in range(0, 1000): #will be updated to reach a certain file size
    
        flag = random.randint(0, 99)
        if (flag == 0): #generate anomaly 1% of the tim
            generateAnomaly(file)
        else:
            line = generateLine()
       
        print line #temp
        #file.write(line)
```

The following function calls each subcomponent of the log and is the main taskmaster of the program

```python
def generateLine(): #this creates a standard line of random variables 
       
        IP = str(generateIP())
        Time = generateTimeStamp() #normal time generation without flag
        Action = generateAction()
        ActionStatus = generateActionStatus()
        Port = str(generatePort())
        User = generateUser()
        return IP + " " + Time+ " " + Action + " " + ActionStatus + " " + Port + " " + User 
```

The following section contains each subcomponent of the log:
IP, Time, Action, Action Status, Port, and User

```python
def generateIP():
    
    global lockIP, lastIP
        
    if (lockIP == 1): #anomaly Scenario triggered
        return lastIP
    else: #normal scenario
        lastIP = socket.inet_ntoa(struct.pack('>I', random.randint(1, 0xffffffff))) #generate random IP in proper format
        return lastIP

def generateTimeStamp(): #function generates time.  Varies between normal and anomaly status

    global Time, lockTime
    
    if (lockTime == 1): #anomaly was triggered
        timeIncrement = datetime.timedelta(seconds = 2) #new request every 2 seconds
    else:  #normalScenario  
        timeIncrement = datetime.timedelta(seconds = random.randint(3, 10000)) #new request every 3 - 10000 seconds
    Time = Time + timeIncrement
    t = Time.strftime('%m/%d/%Y %H:%M:%S')
    return t


def generateAction(): #need to find list of different actions
    return "Generic Action"


def generateActionStatus(): #No good online statuses available
    
    global lockActionStatus
    
    successVariable = random.randint(0, 3)
    
    if (lockActionStatus == 1):
        return "Failure"
    
    else:
        if (successVariable == 0): #25% chance of the access failing
            return "Failure"
        else: 
            return "Success"

def generatePort(): #to be updated to be more realistic

    global lockPort, lastPort
    
    if (lockPort == 1): #portScanning triggered
        lastPort = lastPort+1
        return lastPort
        
    else: #normal scenario    
        lastPort = random.randint(0,100)   
        return lastPort
    
def generateUser():  #this will be updated
    return "User123"
```

The following function is called 1% of the time and chooses to run one out of five potential anomaly scenarios

```python
def generateAnomaly(file):
    
    anomalyType = random.randint(0, 4) #random chance to generate one out of five scenarios
    lines = []
    
    print "ANAMOLY TRIGGERED \n"
    
    if (anomalyType == 0):
       print "DENIAL OF SERVICE \n"
       lines = denialofService()
    
    elif (anomalyType == 1):
        print "repeat"
        lines = repeatLogAttempt()
    
    elif (anomalyType == 2):
        
        print "DENIED REQUEST  \n"
        lines = deniedRequests()
    
    elif (anomalyType == 3):
        print "PORT SCANNING  \n"
        lines = portScanning()
    
    else:
        print "DIFFERENT MACHINES  \n"
        lines = differentMachines()
        
    print "\n".join(lines)
    
    print "ANAMOLY ENDED \n"
```

This following section defines the different types of anomalies:
1)Denial of Service
2)Continuous Requests
3)Repeated Requests
4)Port Scanning
5)Different machine attempts

```python

#scenario where same IP bombards the server
def denialofService():
    
    global lockPort, lockIP, lockTime
    
    #begin locking of Port and IP and double time
    lockPort = 1
    lockIP = 1
    lockTime = 1
    
    
    #make 1000 attempts
    lines = []
    for i in range (0, 1000):
        print generateLine()
        lines.append(generateLine())
    
    
    #end lock
    lockTime = 0
    lockPort = 0
    lockIP = 0
    
    return lines
    
def repeatLogAttempt(): #needs updated functionality where other users log-in in between
    
    #begin lock of IP, User, and set action to "LogIn"
    lockIP = 1
    lockUser = 1
    lockAction = 1
    
    lines = [] #this will hold the lines generated
    for i in range (0, 25):
        lines.append(generateLine())
        
    lockIP = 0
    lockUser = 0
    lockAction = 0
    #end lock    
    
    return lines
    
def deniedRequests():
    global lockIP, lockActionStatus, lastIP
    #begin lock of IP and actionStatus
    lockIP = 1
    lockActionStatus = 1
    
    lines = [] #this will hold the lines generated
    for i in range (0, 25):
        lines.append(generateLine())
    
    #end lock
    lockIP = 0
    lockActionStatus = 0
    
    return lines
        
def portScanning(): #same IP tries to use differnet ports

    global lockPort, lockIP
    #lockIP and cause port to increment by 1
    lockPort = 1
    lockIP = 1
    
    lines = [] 
    
    for i in range (0, 15):
        lines.append(generateLine())
    
    #end lock
    lockPort = 0
    lockIP = 0
    
    return lines
    
def differentMachines(): # to be updated
   
    global lockIP
    
    lockIP = 1
    
    lines = []
    
    for i in range (0, 15):
        lines.append(generateLine())
    
    return lines
    
    lockIP = 0
```
